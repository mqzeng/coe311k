```python
import numpy as np
import matplotlib.pyplot as plt

a, b, N = 0, 2 * np.pi, 100
x = np.linspace(a, b, N)
h = (b - a) / (N - 1)

def f(x):
    return np.sin(x)


df_dx = (f(x + h) - f(x - h)) / (2 * h)
df_forward = (f(x + h) - f(x)) / h
df_backward = (f(x) - f(x - h)) / h

actual_deriv = np.cos(x)

plt.figure(figsize=(20, 12))
plt.plot(x, actual_deriv, label="actual cos(x)")
plt.plot(x, df_forward, label="foward diff")
plt.plot(x, df_backward, label="backward diff")
plt.plot(x, df_dx, label="central diff")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```
