```python
import numpy as np
import time

def jacobi(A, b,iterations = 10000):
    n = len(b)
    x = np.zeros_like(b)
    xnew = np.zeros_like(b)
    
    for i in range(iterations):
        for j in range(n):
            xnew[j] = (b[j] - np.dot(A[j, :], x) + A[j, j]*x[j]) / A[j, j]
        
        if np.linalg.norm(xnew - x, ord=1) < 1e-6:
            return xnew, i+1
        
        x = xnew.copy()
        
    return x, iterations

def gs(A, b, iterations = 10000):
    n = len(b)
    x = np.zeros_like(b)
    

    for i in range(iterations):
        xold = x.copy()
        for j in range(n):
            x[j] = (b[j] - np.dot(A[j, :], x) + A[j, j]*x[j]) / A[j, j]

        if np.linalg.norm(x - xold, ord=1) < 1e-6:
            return x, i+1
    
    return x, iterations

A = np.array([
    [10, -1, 2, 0, 0],
    [-1, 11, -1, 3, 0],
    [2, -1, 10, -1, 0],
    [0, 3, -1, 8, -2],
    [0, 0, 0, -2, 9]
], dtype=float)

b = np.array([14, 30, 26, 25, 37], dtype=float)

start = time.time()
xJac, iterJac = jacobi(A, b)
timeJac = time.time() - start

start = time.time()
xGs, iterGs = gs(A, b)
timeGs = time.time() - start

print("jacobi matrix: ", xJac, "iterations:",iterJac, " time:", timeJac)
print("gs matrix:     ", xGs, "iterations:",iterGs, " time:", timeGs)
print()

C = np.array([
    [1, 2, 3, 0, 0],
    [2, 1, 2, 3, 0],
    [3, 2, 1, 2, 3],
    [0, 3, 2, 1, 2],
    [0, 0, 3, 2, 1]
], dtype=float)

d = np.array([14, 22, 33, 26, 22], dtype=float)

start = time.time()
xJac, iterJac = jacobi(C, d)
timeJac = time.time() - start

start = time.time()
xGs, iterGs = gs(C, d)
timeGs = time.time() - start

print("jacobi matrix: ", xJac, "iterations:",iterJac, " time:", timeJac)
print("gs matrix:     ", xGs, "iterations:",iterGs, " time:", timeGs)

# ^ Solving Cx = d doesn't converge to a single solution --> keeps iterating and solution fails
```
