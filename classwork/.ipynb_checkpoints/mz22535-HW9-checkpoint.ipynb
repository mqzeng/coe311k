```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])
y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

A = np.vstack([x, np.ones(len(x))]).T

m, c = np.linalg.lstsq(A, y, rcond=None)[0]

print(f"slope: {m}, intercept: {c}")

plt.scatter(x, y)
plt.plot(x, m*x + c)
plt.xlabel("x")
plt.ylabel("y")
plt.title("part 1")
plt.show()

np.random.seed(None)

x_rand = np.random.rand(5) * 5
y_rand = np.random.rand(5) * 5

A_rand = np.vstack([x_rand, np.ones(len(x_rand))]).T
m_rand, c_rand = np.linalg.lstsq(A_rand, y_rand, rcond=None)[0]

plt.scatter(x_rand, y_rand)
plt.plot(x_rand, m_rand*x_rand + c_rand)

plt.scatter(x_rand, y_rand)
plt.plot(x_rand, m_rand*x_rand + c_rand)
plt.xlabel("x")
plt.ylabel("y")
plt.title("part 2")
plt.show()

## The least squares regression line still minimizes the squared error,
## but since the data is all random, there really isn't any relationship
## between the data. The line just represents the "best" linear relation

np.random.seed(None)

def trueModel(x):
    return 3*x + 2

x = np.arange(0, 1, 0.01)

noise = np.random.normal(0, 0.1, len(x))
y = trueModel(x) + noise

A = np.vstack([x, np.ones(len(x))]).T
m_fit, c_fit = np.linalg.lstsq(A, y, rcond=None)[0]

print(f"slope: {m_fit}")
print(f"intercept: {c_fit}")

plt.scatter(x, y, s=10)
plt.plot(x, trueModel(x), label="true model")
plt.plot(x, m_fit*x + c_fit, label="fitted model")
plt.legend()
plt.title("original vs fitted")
plt.show()
```
