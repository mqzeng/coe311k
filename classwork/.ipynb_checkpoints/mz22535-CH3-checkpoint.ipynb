```python
import random as rand
import numpy as np
import matplotlib.pyplot as plt

def monteCarlo(samples):
    points = 0
    
    for i in range(samples):
        x = rand.random()
        y = rand.random()
        if x**2 + y**2 <= 1:
            points += 1
    
    return 4 * points / samples

trueErrorList = []
relErrorList = []
run = []

for i in range(100):
    calcPi = monteCarlo(1000)
    pi = np.pi
    trueError = abs(pi - calcPi)
    relError = abs(pi - calcPi) / pi
    
    run.append(i)
    trueErrorList.append(trueError)
    relErrorList.append(relError)

plt.figure(figsize=(15, 6))
plt.plot(run, relErrorList, label = "relative error", marker='o')
plt.plot(run, trueErrorList, label = "true error", marker='s')
plt.xlabel("run")
plt.ylabel("error")
plt.legend()
plt.grid(True)
plt.show()
    
```
