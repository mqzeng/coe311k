```python
def base10Bin(n):
    bin = ''
    print("Quotient     Remainder")
    while n > 0:
            print(n // 2, "          ", n % 2)
        elif len(str(n // 2)) == 3:
            print(n // 2, "           ", n % 2)
        elif len(str(n // 2)) == 2:
            print(n // 2, "            ", n % 2)
        elif len(str(n // 2)) == 1:
            print(n // 2, "             ", n % 2)
            
        bin = str(n % 2) + bin
        n = n // 2
    
    return bin

print(base10Bin(2025))
```
