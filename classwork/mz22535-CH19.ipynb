```python
import numpy as np
import matplotlib.pyplot as plt

step_sizes = [0.1, 0.01, 0.001, 0.00001]

t_exact = np.linspace(0, 30, 10000)
y_exact = 3 - 0.998*np.exp(-1000*t_exact) - 2.002*np.exp(-t_exact)

plt.figure(figsize=(12, 6))
plt.plot(t_exact, y_exact, label="analytical")

for h in step_sizes:
    t = np.arange(0, 30 + h, h)
    y = np.zeros(len(t))
    with np.errstate(over='ignore', invalid='ignore'):
        for i in range(1, len(t)):
            y[i] = y[i-1] + h * (-1000*y[i-1] + 3000 - 2000*np.exp(-t[i-1]))
    plt.plot(t, np.clip(y, -10, 10), label=f"h = {h}")

plt.xlabel("t")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.title("euler's forward vs analytical")
plt.tight_layout()
plt.show()
```
