```python
import numpy as np
import time

def jacobi(A, b,iterations = 10000):
    n = len(b)
    x = np.zeros_like(b)
    xnew = np.zeros_like(b)
    
    for i in range(iterations):
        for j in range(n):
            xnew[j] = (b[j] - np.dot(A[j, :], x) + A[j, j]*x[j]) / A[j, j]
        
        if np.linalg.norm(xnew - x, ord=1) < 1e-6:
            return xnew, i+1
        
        x = xnew.copy()
        
    return x, iterations

def gs(A, b, iterations = 10000):
    n = len(b)
    x = np.zeros_like(b)
    

    for i in range(iterations):
        xold = x.copy()
        for j in range(n):
            x[j] = (b[j] - np.dot(A[j, :], x) + A[j, j]*x[j]) / A[j, j]

        if np.linalg.norm(x - xold, ord=1) < 1e-6:
            return x, i+1
    
    return x, iterations

A = np.array([

[0.582745, 0.48 , 0.10 , 0.  , 0.  ],
[0.48 , 1.044129, 0.46 , 0.10 , 0.  ],
[0.10 , 0.46 , 1.10431 , 0.44 , 0.10 ],
[0.  , 0.10 , 0.44 , 0.963889, 0.42 ],
[0.  , 0.  , 0.10 , 0.42 , 0.522565]], dtype=float)

b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)
             
start = time.time()
xJac, iterJac = jacobi(A, b)
timeJac = time.time() - start

start = time.time()
xGs, iterGs = gs(A, b)
timeGs = time.time() - start

print("jacobi matrix: ", xJac, "iterations:",iterJac, " time:", timeJac)
print("gs matrix:     ", xGs, "iterations:",iterGs, " time:", timeGs)
print()
```
