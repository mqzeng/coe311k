```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 50)
y = np.sin(x)
randData = np.random.randn(500)

plt.figure(figsize=(10,8))
plt.subplot(2, 2, 1)
sizes = [25, 30, 20, 25]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("pie")

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x))
plt.title("line")

plt.subplot(2, 2, 3)
plt.hist(randData, bins=20)
plt.title("histogram")

plt.subplot(2, 2, 4)
plt.scatter(x, y)
plt.title("scatter")

plt.suptitle("4 plots")
plt.show()
```
