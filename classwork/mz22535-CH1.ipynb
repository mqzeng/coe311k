```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 50)
y = np.sin(x)
randData = np.random.randn(500)

plt.figure(figsize=(10,8))
plt.subplot(2, 2, 1)
plt.scatter(x, y)
plt.title("Scatter Plot")

plt.subplot(2, 2, 2)
plt.hist(randData, bins=20)
plt.title("Histogram")

plt.subplot(2, 2, 3)
sizes = [25, 30, 20, 25]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='o')
plt.title("Pie Chart")

plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x))
plt.title("Line Plot")

plt.suptitle("Challenge 1 - 4 Plot Types")
plt.show()
```
