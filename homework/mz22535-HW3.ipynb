```python
import numpy as np
import matplotlib.pyplot as plt
import math

def taylorSin(x, n):
    total = 0
    for i in range(n):
        total += ((-1)**n) * (x**(2*n +1)) / math.factorial(2*n + 1)
    return total

values = np.arange(0, 10.1, 0.1)
sin = np.sin(values)

for i in range(1, 5):
    errors = []
    for j in range(len(values)):
        approx = taylorSin(values[j], i)
        errors.append(abs(sin[i] - approx))
        
    plt.plot(values, errors, label=f"{i} term")

plt.legend()
plt.show()
```
