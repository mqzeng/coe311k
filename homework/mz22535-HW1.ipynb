```python
import random as rand
import math
import matplotlib.pyplot as plt

def monteCarlo(samples):
    points = 0
    
    for i in range(samples):
        x = rand.random()
        y = rand.random()
        if x**2 + y**2 <= 1:
            points += 1
    
    return 4 * points / samples

sizes = range(10, 10000, 20)
errors = []

for i in sizes:
    estimate = monteCarlo(i)
    error = abs(estimate - math.pi)
    errors.append(error)

plt.figure()
plt.plot(sizes, errors, marker = 'o')
plt.xlabel("Sample Number")
plt.ylabel("Error")
plt.grid(True)
plt.show()
```
