```python
import numpy as np
import matplotlib.pyplot as plt

# stability
h_values = [0.01, 0.05, 0.15, 0.20, 0.25, 0.50, 0.75, 1.00]
x = np.linspace(0, 10 * np.pi, 1000)
lam = -2 * np.cos(x)

# plot 1
fig, ax = plt.subplots(figsize=(12, 5))

for h in h_values:
    amp = np.abs(1 + h * lam)
    ax.plot(x / np.pi, amp, label=f'h = {h}')

ax.axhline(1.0, label='stability boundary = 1')
ax.set_title("euler amplification")
ax.set_xlabel("x / π")
ax.set_ylabel("|1 + h·λ|")
ax.legend()
ax.grid()
plt.tight_layout()
plt.show()

# plot 2
stable_fractions = []
for h in h_values:
    amp = np.abs(1 + h * lam)
    stable_fractions.append(np.mean(amp <= 1.0))

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar([str(h) for h in h_values], stable_fractions)
ax.axhline(1.0, label='stable')
ax.set_title("stable fraction interval")
ax.set_xlabel("h")
ax.set_ylabel("stable fraction")
ax.legend()
ax.grid()
plt.tight_layout()
plt.show()

## The bounding conditions do not affect the choice in h, as the stable/unstable split is the same for each h, as the
## amplification factor is only <1 when λ is negative. h simply determines how much that factor is scaled.
```
