```python
import random as rand
import math
import numpy as np
import matplotlib.pyplot as plt

def monteCarlo(n):
    points = 0
    for i in range(n):
        x = rand.random()
        y = rand.random()
        
        if x**2 + y**2 <= i:
            points = points + 1
            
    estimate = 4 * points / n
    return estimate

samples = [10, 100, 1000, 10000]
runs = 10

calcPi = 4 * math.atan(1)
actualPi = np.pi

calcErrors = {}
actualErrors = {}

for i in samples:
    calcErrors[i] = []
    actualErrors[i] = []
    
    for j in range(runs):
        estPi = monteCarlo(i)
        errorCalc = abs(estPi - calcPi)
        errorActual = abs(estPi - actualPi)
        
        calcErrors[i].append(errorCalc)
        actualErrors[i].append(errorActual)
        
for i in samples:
    meanErrorCalc = sum(calcErrors[i]) / runs
    meanErrorActual = sum(actualErrors[i]) / runs
    print("Calculated atan mean error: ", meanErrorCalc)
    print("Actual mean error: ", meanErrorActual)
    
    plt.figure()
    plt.plot(range(1, runs + 1), actualErrors[i], marker = 'o')
    plt.xlabel("Run")
    plt.ylabel("Error")
    plt.title("Sample Size: " + str(i))
    plt.grid(True)
    plt.show()
```
