```python
import numpy as np
import time

def jacobi(A, b,iterations = 10000):
    n = len(b)
    x = np.zeros_like(b)
    xnew = np.zeros_like(b)
    
    for i in range(iterations):
        for j in range(n):
            xnew[j] = (b[j] - np.dot(A[j, :], x) + A[j, j]*x[j]) / A[j, j]
        
        if np.linalg.norm(xnew - x, ord=1) < 1e-6:
            return xnew, i+1
        
        x = xnew.copy()
        
    return x, iterations

def gs(A, b, iterations = 10000):
    n = len(b)
    x = np.zeros_like(b)
    

    for i in range(iterations):
        xold = x.copy()
        for j in range(n):
            x[j] = (b[j] - np.dot(A[j, :], x) + A[j, j]*x[j]) / A[j, j]

        if np.linalg.norm(x - xold, ord=1) < 1e-6:
            return x, i+1
    
    return x, iterations

```
